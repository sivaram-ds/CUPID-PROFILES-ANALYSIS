```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from kmodes.kprototypes import KPrototypes
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.decomposition import PCA

# ✅ Step 1: Ensure Numerical Data is NOT Standardized
# We assume df_cluster already contains raw numerical values (NO StandardScaler applied)

# ✅ Step 2: Convert Categorical Features to Index Values (Required for K-Prototypes)
df_cluster[categorical_cols] = df_cluster[categorical_cols].astype("category")
df_cluster[categorical_cols] = df_cluster[categorical_cols].apply(lambda x: x.cat.codes)

# ✅ Step 3: Apply K-Prototypes Clustering
optimal_k = 4  # Adjust based on Silhouette Score
kproto = KPrototypes(n_clusters=optimal_k, random_state=42, init="Huang", n_init=10)

# Pass correct column indices for categorical data
clusters = kproto.fit_predict(df_cluster, categorical=[df_cluster.columns.get_loc(col) for col in categorical_cols])

# ✅ Assign Cluster Labels
df_cluster["cluster"] = clusters

# ✅ Step 4: Evaluate Clustering Performance
sil_score = silhouette_score(df_cluster[numerical_cols], df_cluster["cluster"])
dbi_score = davies_bouldin_score(df_cluster[numerical_cols], df_cluster["cluster"])

print(f"✅ Silhouette Score (K-Prototypes): {sil_score:.4f}")
print(f"✅ Davies-Bouldin Index (K-Prototypes): {dbi_score:.4f}")

# ✅ Step 5: Save Clustered Data
df_cluster.to_csv("kprototypes_clustered.csv", index=False)
print("✅ Clustered data saved as 'kprototypes_clustered.csv'")

# ✅ Step 6: Visualize Clusters Using PCA (2D & 3D)
pca = PCA(n_components=3)
df_pca = pca.fit_transform(df_cluster[numerical_cols])  # Using actual numerical data

# Convert to DataFrame for plotting
df_pca = pd.DataFrame(df_pca, columns=["PC1", "PC2", "PC3"])
df_pca["Cluster"] = df_cluster["cluster"]

# 🔹 2D PCA Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_pca["PC1"], y=df_pca["PC2"], hue=df_pca["Cluster"], palette="viridis", alpha=0.7)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("Clusters Visualization (2D PCA) - K-Prototypes")
plt.legend(title="Cluster")
plt.show()

# 🔹 3D Interactive PCA Scatter Plot
fig = px.scatter_3d(df_pca, x="PC1", y="PC2", z="PC3", color=df_pca["Cluster"].astype(str),
                     title="3D Cluster Visualization (PCA)", opacity=0.7)
fig.write_html("pca_3d_clusters_kprototypes.html")
print("✅ Interactive 3D PCA visualization saved as 'pca_3d_clusters_kprototypes.html'")
```

```python
import pandas as pd
import spacy
import re
from tqdm import tqdm
from spacy.lang.en.stop_words import STOP_WORDS

# ✅ Load Spacy Model (Using Small Model for Speed)
nlp = spacy.load("en_core_web_sm", disable=["ner", "parser"])

# ✅ Load Essay Dataset
df_nlp = pd.read_csv("profiles_essays.csv")

# ✅ Combine All Essays into One Column
df_nlp["all_essays"] = df_nlp[
    ["essay0", "essay1", "essay2", "essay3", "essay4",
     "essay5", "essay6", "essay7", "essay8", "essay9"]
].astype(str).agg(" ".join, axis=1)

# ✅ **Fix Read-Only Buffer** (Force Copying Text Data)
df_nlp["all_essays"] = df_nlp["all_essays"].copy()

# ✅ Text Preprocessing Function
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)  # Remove URLs
    text = re.sub(r"\W", " ", text)  # Remove special characters
    text = re.sub(r"\d+", "", text)  # Remove numbers

    doc = nlp(text)  # Process text with Spacy
    tokens = [token.lemma_ for token in doc if token.text not in STOP_WORDS and not token.is_punct]

    return " ".join(tokens)

# ✅ Apply Preprocessing with Progress Bar (Sequential Processing)
tqdm.pandas()
df_nlp["processed_essays"] = df_nlp["all_essays"].progress_apply(preprocess_text)

# ✅ Remove Empty Essays
df_nlp = df_nlp[df_nlp["processed_essays"].str.len() > 10]

# ✅ Save Cleaned Essays
df_nlp[["processed_essays"]].to_csv("cleaned_essays.csv", index=False)

print("✅ Essay Preprocessing Completed Successfully!")
```

```python
# Convert text to embeddings (768-dimensional vectors)
print("⚡ Converting Essays into Vector Embeddings...")
embeddings = model.encode(essays, batch_size=64, show_progress_bar=True, device=device)

# Convert to NumPy array
embeddings = np.array(embeddings)

# Save embeddings for future use
np.save("profile_embeddings.npy", embeddings)

print(f"✅ Embeddings Shape: {embeddings.shape}")  # Should be (num_profiles, 768)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load saved embeddings
print("⚡ Loading Saved Embeddings...")
embeddings = np.load("profile_embeddings.npy")

# Define batch size
batch_size = 500  # Reduce further if needed
num_profiles = len(embeddings)

# File to save results
similarity_file = "similarity_matrix_batches.npy"

# Open a file to save results incrementally
with open(similarity_file, "wb") as f:
    print(f"⚡ Computing & Saving Similarity Scores in Batches (Batch Size: {batch_size})...")

    for start in range(0, num_profiles, batch_size):
        end = min(start + batch_size, num_profiles)
        print(f"Processing Batch: {start} - {end}...")

        # Compute similarity for this batch
        batch_similarities = cosine_similarity(embeddings[start:end], embeddings)

        # Save batch to file
        np.save(f, batch_similarities)

print("✅ Similarity Computation Completed & Saved in Batches!")
```

```python
import numpy as np
import pandas as pd

# ✅ Load Similarity Matrix (Ensure Correct File Name)
print("⚡ Loading Similarity Scores...")
try:
    similarity_matrix = np.load("similarity_matrix_batches.npy", allow_pickle=True)
    print("✅ Similarity Matrix Loaded Successfully!")
except FileNotFoundError:
    print("❌ ERROR: Similarity Matrix File Not Found!")
    exit()

# ✅ Load Clustered Data
print("⚡ Loading Clustered Data...")
try:
 df_cluster = pd.read_csv("kprototypes_clustered.csv")
 print("✅ Clustered Data Loaded Successfully!")
except FileNotFoundError:
    print("❌ ERROR: Clustered Data File Not Found!")
    exit()

# ✅ Load Processed NLP Essays
print("⚡ Loading Processed Essays...")
try:
    df_nlp = pd.read_csv("cleaned_essays.csv")
    print("✅ Processed Essays Loaded Successfully!")
except FileNotFoundError:
    print("❌ ERROR: Processed Essays File Not Found!")
    exit()

# ✅ Set Number of Matches
top_n = 10

# ✅ Dictionary to Store Recommendations
recommendations = {}

print("⚡ Generating Top 10 Matches Per Profile...")
for i in range(len(similarity_matrix)):
    # Get similarity scores for profile i
    sim_scores = similarity_matrix[i]

    # Rank profiles by similarity (descending)
    top_matches = np.argsort(sim_scores)[::-1][1:top_n+1]  # Exclude self (index 0)

    # Get only matches from the same cluster
    same_cluster_matches = [idx for idx in top_matches if df_cluster.loc[idx, "cluster"] == df_cluster.loc[i, "cluster"]]

    # Store matches
    recommendations[i] = same_cluster_matches[:top_n]  # Ensure max 10 matches

# ✅ Save Recommendations
recommendations_df = pd.DataFrame.from_dict(recommendations, orient="index")
recommendations_df.to_csv("final_recommendations.csv", index=False)

print("✅ Final Matchmaking Completed & Saved as 'final_recommendations.csv'!")
```

```python
import dash
from dash import dcc, html, Input, Output
import pandas as pd
import plotly.express as px
import dash_cytoscape as cyto
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import base64
import io

# ✅ Load Data
df_cluster = pd.read_csv("kprototypes_clustered.csv")
df_nlp = pd.read_csv("profiles_essays_fixed.csv", dtype=str, low_memory=False)

# Ensure cluster is categorical
df_cluster["cluster"] = df_cluster["cluster"].astype(str)

# ✅ Features Available for Cluster Insights
cluster_features = ["age", "height", "income", "sex", "status", "location", "drinks", "smokes", "pets", "sign", "body_type", "diet",
    "orientation", "education", "drugs", "ethnicity", "job", "offspring", "religion"]

# ✅ Generate Word Cloud Images for NLP Clusters (Ignoring NaN)
def generate_wordcloud(cluster):
    cluster_texts = df_nlp.loc[df_cluster["cluster"] == cluster, "processed_essays"].dropna()
    text = " ".join(cluster_texts)

    # Remove NaN-related words & trim whitespace
    text = text.replace("nan", "").strip()

    if not text:  # If empty after cleaning, return blank
        return ""

    wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)

    img_buffer = io.BytesIO()
    wordcloud.to_image().save(img_buffer, format="PNG")
    img_buffer.seek(0)

    encoded_img = base64.b64encode(img_buffer.read()).decode()
    return f"data:image/png;base64,{encoded_img}"

wordcloud_images = {cluster: generate_wordcloud(cluster) for cluster in df_cluster["cluster"].unique()}
# ✅ Get Categorical Mappings
category_mappings = {}
for col in categorical_cols:
    category_mappings[col] = dict(enumerate(df_cluster[col].astype("category").cat.categories))

# ✅ Replace numeric values with actual labels
for col in categorical_cols:
    df_cluster[col] = df_cluster[col].map(category_mappings[col])

# ✅ Dash App Layout
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Matchmaking Project Dashboard", style={"textAlign": "center"}),

    # Dropdowns for Cluster & Feature Selection
    html.Label("Select Cluster:"),
    dcc.Dropdown(
        id="cluster-dropdown",
        options=[{"label": f"Cluster {c}", "value": c} for c in df_cluster["cluster"].unique()],
        placeholder="Select a Cluster",
    ),

    html.Label("Select Feature for Insights:"),
    dcc.Dropdown(
        id="feature-dropdown",
        options=[{"label": feature.capitalize(), "value": feature} for feature in cluster_features],
        placeholder="Select a Feature",
    ),

    # Cluster Feature Distribution Chart
    dcc.Graph(id="cluster-feature-distribution"),

    # Word Cloud for NLP (Hide if Empty)
    html.Div([
        html.H3("Word Cloud for NLP Cluster"),
        html.Img(id="wordcloud", style={"width": "80%", "display": "block", "margin": "auto"})
    ], id="wordcloud-container", style={"display": "none"}),


])

# ✅ Callback for Cluster Feature Distribution
# ✅ Callback for Cluster Feature Distribution
@app.callback(
    Output("cluster-feature-distribution", "figure"),
    [Input("cluster-dropdown", "value"),
     Input("feature-dropdown", "value")]
)
def update_cluster_feature_distribution(selected_cluster, selected_feature):
    if not selected_feature:
        return px.histogram(df_cluster, x="cluster", title="Cluster Distribution")

    # ✅ Apply categorical label mapping if selected_feature is categorical
    if selected_feature in category_mappings:
        df_cluster[selected_feature] = df_cluster[selected_feature].replace(category_mappings[selected_feature])

    # ✅ Filter data for the selected cluster
    if selected_cluster:
        cluster_data = df_cluster[df_cluster["cluster"] == selected_cluster]
        fig = px.histogram(cluster_data, x=selected_feature, title=f"{selected_feature.capitalize()} Distribution for Cluster {selected_cluster}")
    else:
        fig = px.histogram(df_cluster, x=selected_feature, color="cluster", barmode="overlay",
                           title=f"{selected_feature.capitalize()} Distribution Across Clusters")

    return fig

# ✅ Callback for Word Cloud Updates (Hide if No Data)
@app.callback(
    [Output("wordcloud", "src"), Output("wordcloud-container", "style")],
    Input("cluster-dropdown", "value")
)
def update_wordcloud(selected_cluster):
    if selected_cluster in wordcloud_images and wordcloud_images[selected_cluster]:
        return wordcloud_images[selected_cluster], {"display": "block"}
    return "", {"display": "none"}  # Hide if empty

# ✅ Run the App
if __name__ == "__main__":
    app.run(debug=True, host="0.0.0.0", port=8060)


```
